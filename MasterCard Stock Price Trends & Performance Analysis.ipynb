```python
import pandas as pd

# Load stock history data
Data_Path = "C:\\Users\\HP\\Downloads\\archive\\"

Data_File = "Mastercard_stock_history.csv"

df =df = pd.read_csv(Data_Path + Data_File, parse_dates=["Date"], index_col="Date")


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y', errors='coerce')


# Plot stock closing price over time
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='Date', y='Close', label="Closing Price", color="green")

plt.title("Mastercard Stock Closing Price Trend")
plt.xlabel("Date")
plt.ylabel("Stock Price ($)")
plt.legend()
plt.show()
```

```python

# Ensure the data is sorted by date
df = df.sort_index()

# Compute Moving Averages
df["SMA_50"] = df["Close"].rolling(window=50).mean()   # 50-day SMA
df["SMA_200"] = df["Close"].rolling(window=200).mean() # 200-day SMA
df["EMA_20"] = df["Close"].ewm(span=20, adjust=False).mean()  # 20-day EMA

# Plot Closing Price with Moving Averages
plt.figure(figsize=(13,7))
plt.plot(df["Close"], label="Closing Price", color="green", linewidth=1)
plt.plot(df["SMA_50"], label="50-day SMA", color="blue", linestyle="dashed")
plt.plot(df["SMA_200"], label="200-day SMA", color="red", linestyle="dashed")
plt.plot(df["EMA_20"], label="20-day EMA", color="orange")

# Chart Enhancements
plt.title("Mastercard Stock Price with Moving Averages")
plt.xlabel("Date")
plt.ylabel("Stock Price ($)")
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

```python
import numpy as np

# Calculate daily returns
df["Daily Return"] = df["Close"].pct_change()

# Calculate stock volatility (std deviation of daily returns)
volatility = np.std(df["Daily Return"]) * 100
print(f"Stock Volatility: {volatility:.2f}%")
```
